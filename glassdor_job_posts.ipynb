# Clean and format data frame
## To Do:

1.   Set new index.
2.   Remove numbers from the company name column.
4.   Separate location column into state and city.
5.   Create new *company age* column based on the year founded.
6.   Create new *max_salary*, *min_salary* and *avg_salary* columns.
7.   Clean *estimate salary* column.
8.   Save new data set into a new csv file.



```python
clean_df = df.copy()
```

```python
# Set new index
clean_df.set_index('index', inplace=True)
```

```python
clean_df.head()
```

```python
# Clean company name column
clean_df['Company Name'] = clean_df['Company Name'].apply(lambda x: x.split('\n')[0])
```

```python
clean_df['Company Name']
```

```python
# Separate location into City and State
clean_df['City'] = clean_df['Location'].apply(lambda x: x.split(', ')[0])
clean_df['State'] = clean_df['Location'].apply(lambda x: x.split(", ")[-1])
```

```python
# Drop Location column
clean_df.drop('Location', axis=1, inplace=True)
```

```python
# Show changes
clean_df.head()
```

```python
# Format data type of the 'Founded' column.
clean_df['Founded'] = clean_df['Founded'].astype('int64')
```

```python
# Create new column with the company age
clean_df['Company Age'] = clean_df['Founded'].apply(lambda x: 2023 - x)
```

```python
# Show changes
clean_df.head()
```

```python
# Clean 'Salary Estimate' column. Remove 'K' and '$'.
clean_df['Salary Estimate'] = clean_df['Salary Estimate'].apply(lambda x: x.split('(')[0].replace('$', '').replace('K', ''))
```

```python
# Show changes
clean_df['Salary Estimate']
```

```python
# Create new columns:
#   1. Minimum salary
#   2. Maximum salary
#   3. Average salary

clean_df['Min Salary'] = clean_df['Salary Estimate'].apply(lambda x: x.split('-')[0]).astype('int64')
clean_df['Max Salary'] = clean_df['Salary Estimate'].apply(lambda x: x.split('-')[-1]).astype('int64')
clean_df['Avg Salary'] = np.mean(clean_df[['Min Salary', 'Max Salary']], axis=1)
```

```python
clean_df.head()
```

```python
# Save cleaned data set
clean_df.to_csv('clean_DS_jobs.csv')
```
